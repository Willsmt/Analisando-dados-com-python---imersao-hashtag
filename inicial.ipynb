```python
#passo 1 - importar a biblioteca pandas

import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")



#passo 2 - visualizar a tabela (banco de dados)
#informações que não te ajudam, te atrapalham
tabela = tabela.drop(columns="CustomerID")
display(tabela) 
#passo 3 - corrigir os problemas da base de dados
# linhas duplicadas
# valores em formatos errados
# colunas desnecessárias
# valores faltando
```

```python
display(tabela.info())
#excluir linhnas varias dropna  nan- not a number = vazio
tabela = tabela.dropna()

display(tabela.info())
```

```python
# passo 4: analise inicial (entender quantos clientes cancelaram)

#quantidade - contar quantos caras são 0 e quantos são 1 na coluna cancelou
display(tabela["cancelou"].value_counts())

# porcentagem

display(tabela["cancelou"].value_counts(normalize=True)) #só fazer formatação numerica no final do projeto.

#display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# passo 5: análise detalhada (causa do cancelamento dos clientes, cruzar informações)

import plotly.express as px
print(tabela.columns) #tabela.columns mostra todas as colunas da tabela
# primeiro voce cria o grafico e depois voce exibe o grafico
#criar grafico
for coluna in tabela.columns:
    grafico= px.histogram(tabela, x= coluna, color="cancelou", barmode="group",text_auto=True)

    #exibir grafico
    grafico.show()
```

```python
#1 duracao_contrato
# todo mundo do contrato mensal, cancelou o serviço
# quase ninguem do contrato anual cancelou o serviço
# contrato semestral tem uma taxa de cancelamento intermediaria
# dar desconto para o contrato anual

#2 ligações_callcenter  acima de 4 , todo mundo cancelou

# tem algum problema que não estamos conseguindo resolver
#se o cliente ligar 3x pro call center, ele tem mais chances de cancelar o serviço(alerta vermelho ter um time especifico para atender esses clientes)

#3 atrasos acime de 20 dias o cliente cancelca

# politica se o cliente atrasar 15 dias no pagamento, alerta vermelho time especifico para atender esses clientes do financeiro.

#cancelamento = 56%

#filtrar base
#duracao_contrato -> diferente de Monthly
for coluna in tabela.columns:
    if coluna == "duracao_contrato":
        condicao = tabela["duracao_contrato"] != "Monthly"
        tabela = tabela[condicao]
#é a mesma coisa tabela = tabela[tabela["duracao_contrato"] != "Monthly"]
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

#ligacoes_callcenter -> menores ou iguais a 4 
for coluna in tabela.columns:
    if coluna == "ligacoes_callcenter":
        condicao = tabela["ligacoes_callcenter"] <= 4
        tabela = tabela[condicao]
#é a mesma coisa tabela = tabela[tabela["ligacoes_callcenter"] <= 4]
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

# dias_atraso -> menores ou iguais a 20
for coluna in tabela.columns:
    if coluna == "dias_atraso":
        condicao = tabela["dias_atraso"] <= 20
        tabela = tabela[condicao]
# é a mesma coisa tabela = tabela[tabela["dias_atraso"] <= 20]

#porcentagem
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))


```
